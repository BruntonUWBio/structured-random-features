```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt
import pickle
```

## initialize MNIST

```python
h, lr, n_epochs = 1000, 0.1, 3001
t, l = 5, 2
models = ['V1', 'He']
labels = {'V1': 'V1-inspired', 'He': 'standard'}
with open('results/initialize_mnist/full_data/clf_t=%0.2f_l=%0.2f.pickle' % (t, l), 'rb') as handle:
    sims = pickle.load(handle)

colors = {'V1': '#2c7fb8', 'He': '#d95f02'}
fig = plt.figure(figsize=(10.6, 8))
ax = fig.add_subplot(111)
for m in models:
    avg_test_err = np.mean(sims['test_err'][m][h][lr], axis=0)
    std_test_err = np.std(sims['test_err'][m][h][lr], axis=0)
    plt.plot(np.arange(n_epochs), avg_test_err, label=labels[m], lw=7, c=colors[m])
    plt.fill_between(np.arange(n_epochs), avg_test_err - std_test_err, avg_test_err + std_test_err, alpha=0.2)
    print('%s, %0.4f' % (m, avg_test_err[-1] * 100))

plt.xlabel('epochs', fontsize=40)
plt.ylabel('Classification error', fontsize=40)
plt.yticks(np.arange(0, 1.01, 0.1))
plt.xticks(np.arange(0, 3001, 1000))
plt.xlim([0, 3001])
plt.ylim([-0.03, 0.5])
ax.tick_params(axis = 'both', which = 'major', labelsize = 40, width=2, length=6)
# plt.legend(loc = 'upper right', fontsize=30)
plt.savefig('results/paper_figs/figure_7/MNIST_init_results.pdf')
```

## initialize kMNIST

```python
h, lr, n_epochs = 1000, 0.1, 3001
models = ['V1', 'He', 'mistuned']
labels = {'V1': 'V1-inspired', 'He': 'standard', 'mistuned': 'mistuned'}

# load good params
t, l = 5, 2
with open('results/initialize_kmnist/full_data/clf_t=%0.2f_l=%0.2f.pickle' % (t, l), 'rb') as handle:
    results = pickle.load(handle)
    
# load bad params
t, l = 0.5, 0.5
with open('results/initialize_kmnist/full_data/clf_t=%0.2f_l=%0.2f.pickle' % (t, l), 'rb') as handle:
    results_mistuned = pickle.load(handle)

results['test_err']['mistuned'] = results_mistuned['test_err']['V1']

colors = {'V1': '#2c7fb8', 'He': '#d95f02', 'mistuned': '#91cf60'}
fig = plt.figure(figsize=(10.6, 8))
ax = fig.add_subplot(111)
for m in models:
    avg_test_err = np.mean(results['test_err'][m][h][lr], axis=0)
    std_test_err = np.std(results['test_err'][m][h][lr], axis=0)
    plt.plot(np.arange(n_epochs), avg_test_err, label=labels[m], lw=7, c=colors[m])
    plt.fill_between(np.arange(n_epochs), avg_test_err - std_test_err, avg_test_err + std_test_err, alpha=0.2)
    print('%s, %0.4f' % (m, avg_test_err[-1] * 100))

plt.xlabel('epochs', fontsize=40)
plt.ylabel('Classification error', fontsize=40)
plt.yticks(np.arange(0, 1.01, 0.1))
plt.xticks(np.arange(0, 3001, 1000))
plt.xlim([0, 3001])
plt.ylim([-0.03, 0.5])
ax.tick_params(axis = 'both', which = 'major', labelsize = 40, width=2, length=6)
# plt.legend(loc = 'upper right', fontsize=30)
# plt.yscale('log')
plt.savefig('results/paper_figs/figure_7/kMNIST_init_results.pdf')
```
