01/28/21: Scripts to plot Figure 2 for the random features paper. This figure deals withe examples of vision inspired GPs. 

```python
import numpy as np
import matplotlib.pyplot as plt 
from estimator import V1_inspired_weights, V1_inspired_weights_for_center
from palettable.colorbrewer.diverging import PuOr_9, BrBG_7, RdBu_11, PiYG_7
from palettable.cmocean.diverging import Balance_5
```

## Sampled receptive fields from V1

```python
M, N, t, l = 9, 784, 5, 2
dim = int(np.sqrt(N))
W = V1_inspired_weights(M, N, t, l, scale=1, random_state=8) # 784(8)

fig = plt.figure(figsize=(4, 2))
for i in range(3):
    ax = fig.add_subplot(1, 3, i + 1)
    plt.imshow(W[i].reshape(dim, dim), cmap=RdBu_11.mpl_colormap)
    cmap_lim = np.max(np.abs(W[i]))
    plt.clim(-cmap_lim, cmap_lim)
    ax.tick_params(axis='both', which='both', left=False, bottom=False, labelleft=False, labelbottom=False)
#     plt.axis('off')
cbar = plt.colorbar(extend='neither', spacing='proportional',
                orientation='vertical', aspect=8)
cbar.ax.tick_params(axis='y', which='both', length=0, labelright=False)
plt.subplots_adjust(wspace=0.1, hspace=0)
plt.savefig('results/paper_figs/figure_2/V1_weights.pdf')
```

## Change in RF size

```python
M, N, center = 9, 784, [14, 14]
size_params_list = [1, 2.5, 4.5]
freq_param = 3.5

for size_param in size_params_list:
    fig = plt.figure(figsize=(4, 2))
    for i in range(3):   
        ax = fig.add_subplot(1, 3, i + 1)
        W = V1_inspired_weights_for_center(N, size_param, freq_param, center)
        plt.imshow(W.reshape(28, 28), cmap=RdBu_11.mpl_colormap)
        cmap_lim = np.max(np.abs(W))
        plt.clim(-cmap_lim, cmap_lim)
        ax.tick_params(axis='both', which='both', left=False, bottom=False, labelleft=False, labelbottom=False)
    plt.subplots_adjust(wspace=0.1, hspace=0)
#     plt.savefig('results/paper_figs/figure_2/V1_weights_size=%0.2f_freq=%0.2f.pdf' % (size_param, freq_param))
```

```python
M, N, center = 9, 784, [14, 14]
size_params_list = [1, 1.5, 2.7, 4, 5, 8]
freq_param = 4

fig = plt.figure(figsize=(8, 6))
for i, size_param in enumerate(size_params_list):   
    ax = fig.add_subplot(1, 6, i + 1)
    W = V1_inspired_weights_for_center(N, size_param, freq_param, center, random_state=2)
    plt.imshow(W.reshape(28, 28), cmap=RdBu_11.mpl_colormap)
    cmap_lim = np.max(np.abs(W))
    plt.clim(-cmap_lim, cmap_lim)
    ax.tick_params(axis='both', which='both', left=False, bottom=False, labelleft=False, labelbottom=False)
# plt.subplots_adjust(wspace=0.1, hspace=0)
plt.savefig('results/paper_figs/figure_2/V1_weights_increasing_size.pdf')
```
