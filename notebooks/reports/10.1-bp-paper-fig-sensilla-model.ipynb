```python
import os.path as path
from os import getcwd

import matplotlib.pyplot as plt

from src.data.load_dataset import load_sensilla_sta
from src.models.weights import sensilla_weights
```

```python
data_dir = path.abspath(path.join(getcwd(), '../../'))
```

```python
## plot the STA from Pratt
sta, samp_freq, duration = load_sensilla_sta()

fig = plt.figure(figsize=(6, 3))
plt.plot(sta[48], lw=5, c="#047495")
plt.plot(sta[22], lw=5, c="#047495", alpha=0.4)
# plt.axis('off')
plt.savefig(data_dir + '/reports/figures/paper_figs/figure_2/biological_sta.pdf', 
            dpi=300, transparent=False, 
            bbox_inches = 'tight', pad_inches = 0)
```

```python
## plot a model that looks similar to Pratt's
omega_a, omega_b, gamma = 3, 8, 6
W = sensilla_weights(20, 150, omega_a, omega_b, gamma, seed=0)
fig = plt.figure()
plt.plot(W[3], lw=5, c='#e50000')
plt.plot(W[5], lw=5, c='#e50000', alpha=0.5)
plt.axis('off')
plt.savefig(data_dir + '/reports/figures/paper_figs/figure_2/generated_sta.pdf', 
            dpi=300, transparent=True,
           bbox_inches = 'tight', pad_inches = 0)
```

```python
## plot weights with increasing stim frequency
omega_a, gamma = 3, 6
omega_b_list = [5, 20, 100]

fig = plt.figure(figsize=(6, 10))
for i, omega_b in enumerate(omega_b_list):
    W = sensilla_weights(20, 150, omega_a, omega_b, gamma, seed=10)
    ax = fig.add_subplot(3, 1, i + 1)
    plt.plot(W[12], lw=5, c='#e50000')
    plt.axis('off')
plt.savefig(data_dir + '/reports/figures/paper_figs/figure_2/tune_smoothness_sta.pdf', 
            dpi=300, transparent=True,
            bbox_inches = 'tight', pad_inches = 0)
```
